```python
#Não temos uma coluna content type mas é possivel criar ela
df[['acao', 'evento']] = df['evento'].str.split(":",expand=True)
```

```python
# Lista ordenada das etapas com filtros
etapas = [
    ('Instalacao', (df['acao'] == 'visualizacao_tela') & (df['evento'] == 'instalacao')),
    ('Criar Conta', (df['acao'] == 'clique_botao') & (df['evento'] == 'criar_conta')),
    ('Cadastro', (df['acao'] == 'clique_botao') & (df['evento'] == 'cadastro')),
    ('PIN Cadastro', (df['acao'] == 'clique_botao') & (df['evento'] == 'pin_cadastro')),
    ('Inicio App', (df['acao'] == 'visualizacao_tela') & (df['evento'] == 'inicio')),
    ('Ler Minibook', (df['acao'] == 'clique_botao') & (df['evento'] == 'ler_minibook')),
    ('Ler Ebook', (df['acao'] == 'clique_botao') & (df['evento'] == 'ler_ebook')),
    ('Ouvir Audiobook', (df['acao'] == 'clique_botao') & (df['evento'] == 'ouvir_audiobook'))
]

# Inicializar
usuarios_anteriores = set()
funil_filtrado = []

for i, (etapa_nome, filtro) in enumerate(etapas):
    usuarios_etapa = set(df.loc[filtro, 'usuario_id'])

    # Na primeira etapa, pega todos os usuários
    if i == 0:
        usuarios_filtrados = usuarios_etapa
    else:
        usuarios_filtrados = usuarios_etapa & usuarios_anteriores

    funil_filtrado.append({
        'etapa': etapa_nome,
        'usuarios_unicos': len(usuarios_filtrados)
    })

    usuarios_anteriores = usuarios_filtrados

# Converter para DataFrame e calcular taxa de conversão
df_funil_seq = pd.DataFrame(funil_filtrado)
df_funil_seq['taxa_conversao'] = df_funil_seq['usuarios_unicos'] / df_funil_seq['usuarios_unicos'].shift(1)
df_funil_seq['taxa_conversao'] = df_funil_seq['taxa_conversao'].fillna(1).apply(lambda x: f"{x:.2%}")

# Exibir resultado
df_funil_seq
```

```python


# Criar gráfico funil
fig = px.funnel(
    df_funil_seq,
    y='etapa',
    x='usuarios_unicos',
    text='taxa_conversao',
    title='Funil de Cadastro e Consumo - Sequencial',
    labels={'usuarios_unicos': 'Usuários Únicos', 'etapa': 'Etapas'}
)
fig.update_traces(textposition='inside', textfont_size=14)
fig.show()
```
